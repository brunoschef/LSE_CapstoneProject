```python
from pydub import AudioSegment, silence
import os

min_silence_len = 900  # minimum length of silence to split on in milliseconds
keep_silence = 100  # amount of silence to keep at the beginning and end of each clip

# Directory to save the utterances
output_dir = os.path.join(os.getcwd(), "wav_utterances")

# Create the directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

for file in file_list:
    # Load audio file
    audio = AudioSegment.from_file(file, format="wav")

    # Calculate silence threshold
    silence_thresh = audio.dBFS - 14

    # Detect silence
    chunks = silence.split_on_silence(audio, min_silence_len=min_silence_len, 
                                      silence_thresh=silence_thresh, keep_silence=keep_silence)
    
    # Export each chunk with new names
    base_name = os.path.splitext(os.path.basename(file))[0]  # get the base name of the file without the directory path
    for i, chunk in enumerate(chunks):
        new_file_name = f"{base_name}_utterance{i+1}.wav"
        output_path = os.path.join(output_dir, new_file_name)
        chunk.export(output_path, format="wav")

    print(f"Processed {file}: Total utterances: {len(chunks)}")
```

```python
# Change each file to mono audio and 16 kHz sample rate
import os
from pydub import AudioSegment

# Specify the directory containing the audio files
input_dir = "/Users/willscheffler/Desktop/LSE/MY498/wavfiles/wav_utterances"
output_dir = "/Users/willscheffler/Desktop/LSE/MY498/wavfiles/wav_utterances_processed"

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Iterate over all files in the input directory
for file in os.listdir(input_dir):
    # Check if the file is a WAV file
    if file.endswith(".wav"):
        # Load the audio file
        audio_path = os.path.join(input_dir, file)
        audio = AudioSegment.from_file(audio_path, format="wav")
        
        # Convert to mono and resample to 16 kHz
        audio = audio.set_channels(1)
        audio = audio.set_frame_rate(16000)
        
        # Export the processed audio to the output directory with the same filename
        output_path = os.path.join(output_dir, file)
        audio.export(output_path, format="wav")

print("Audio processing complete.")
```
